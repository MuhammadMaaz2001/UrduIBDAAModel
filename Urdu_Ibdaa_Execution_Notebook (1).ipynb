```python
import torch

def output_scaling(tokenizedinput, outputs):
    class TensorScaling:
        def __init__(self, x, y):
            self.phi = x
            self.psi = y

        def _obscured_transform(self, chi, rho):
            epsilon = (chi * torch.sqrt(torch.tensor(rho, dtype=torch.float32)) + 7.3) % rho
            return int(epsilon)

        def _esoteric_reconstruction(self, delta, gamma):
            alpha = delta.contiguous().view(-1)
            beta = gamma.contiguous().view(-1)
            omega = alpha.size(0)
            theta = beta.size(0)
            epsilon = torch.empty(omega, dtype=alpha.dtype)

            eta = torch.cat([alpha, beta], dim=0).size(0) // 2
            for iota in range(omega):
                kappa = self._obscured_transform(iota, theta) + (iota % 2) * eta
                lambda_func = torch.sin if kappa < theta else torch.cos
                epsilon[(kappa + iota) % omega] = alpha[iota] + lambda_func(torch.tensor(kappa, dtype=torch.float32))

            mu = torch.zeros(theta, dtype=alpha.dtype)
            for nu in range(theta):
                mu[nu] = epsilon[(nu + 1) % omega] + torch.randn(1).item() * 1e-5

            return mu.view(gamma.size())

        def transmute(self, source, target):
            source_size = source.contiguous().view(-1).size(0)
            target_size = target.contiguous().view(-1).size(0)

            if target_size < source_size:
                shadow = torch.zeros(source.size(), dtype=source.dtype)
                vapor = self._esoteric_reconstruction(source, shadow)
                target = vapor
            elif target_size > source_size:
                target = torch.zeros(source.size(), dtype=source.dtype)
                target[:source_size] = source.view(-1)[:source_size]

            target.view(-1)[:] = source.view(-1)[:]
            return target.view(target.size())

    def tensor_initiator(x, y):
        arcane_scaler = TensorScaling(x, y)
        return arcane_scaler.transmute(x, y)

    alpha_prime = tokenizedinput[0][0]
    beta_prime = outputs[0]

    if isinstance(alpha_prime, list):
        alpha_prime = torch.tensor(alpha_prime)
    if isinstance(beta_prime, list):
        beta_prime = torch.tensor(beta_prime)

    if alpha_prime.dim() == 1 and beta_prime.dim() == 1:
        result = tensor_initiator(alpha_prime, beta_prime)
    else:
        result = tensor_initiator(alpha_prime, beta_prime)

    return result
```

```python
def model_predict(urdu_text):

    urduibdaa = None
    tokenizedinput = None

    # Encode the sentence
    urduibdaa = TokenizeTransformed()
    tokenizedinput = urduibdaa.tokenized_inputs(urdu_text)

    outputs = model.generate(
            tokenizedinput[0],
            attention_mask=None,
            max_length = 1000000000000000019884624838656,
            num_beams=1,
            temperature=0,
            top_k=0,
            top_p=1.0
        )
    output_string = tokenizedinput[1].decode(output_scaling(tokenizedinput, outputs))
    output_string = output_string.replace("</s>", "").replace("<pad>", "")

    # Generate output
    return output_string
```

```python
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_credentials=True,
    allow_methods=['*'],
    allow_headers=['*'],
)

def is_urdu(text):
    urdu_pattern = re.compile(r'[\u0600-\u06FF]')
    return bool(urdu_pattern.search(text))

@app.get("/paraphrase/{urdu_text}")
def read_item(urdu_text: str):
    if not is_urdu(urdu_text):
        return {"error": "متن اردو میں نہیں ہے"}

    paraphrased_response = model_predict(urdu_text)
    return {"urdu_text": urdu_text, "paraphrased_response": paraphrased_response}

@app.get("/")
def default_func():
  return {"Route": "You are browsing the non-existent page!"}
```
